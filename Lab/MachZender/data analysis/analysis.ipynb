```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
from scipy.optimize import curve_fit
```

```python
df1 = pd.read_excel('../data/Exp05_Mach_Zehnder.xlsx',"Sheet1")
```

```python
print(df1.columns)
```

```python
print(df1['Pressure(mm Hg)'])
print(df1['No of fringes shifted '])
```

```python
pressure_arr_1 = df1['Pressure(mm Hg)'].to_numpy()
fring_arr_1 = df1['No of fringes shifted '].to_numpy()
```

```python
def stLine(x, p1, p2):
    return p1 * x + p2
```

```python
plt.figure(figsize=(10,8))

# Scatter plot
plt.scatter(pressure_arr_1, fring_arr_1)

# Curve fitting
params, pcov = curve_fit(stLine, pressure_arr_1, fring_arr_1)
perr = np.sqrt(np.diag(pcov))

# Predicted values
y_pred = stLine(pressure_arr_1, *params)

# R^2 calculation
ss_res = np.sum((fring_arr_1 - y_pred)**2)
ss_tot = np.sum((fring_arr_1 - np.mean(fring_arr_1))**2)
r_squared = 1 - (ss_res / ss_tot)

# Plot fitted line
plt.plot(
    pressure_arr_1,
    stLine(pressure_arr_1, *params),
    color='orange',
    label=(
        f"Linear Fit: $f(x)=mx+c$\n"
        f"$m = {params[0]:.4f} \\pm {perr[0]:.1e}, "
        f"c = {params[1]:.1f} \\pm {perr[1]:.1f}$\n"
        f"$R^2 = {r_squared:.4f}$"
    )
)

# Labels and styling
plt.xlabel("Pressure Applied (mm Hg)", fontsize=16)
plt.ylabel(r"Number of fringe shift", math_fontfamily='cm', fontsize=16)
plt.title(r"Variation of Fringe shift over pressure change", fontsize=18, fontweight='bold')

plt.legend(loc="best")
plt.grid(True)

# Save and show
plt.savefig('Plot1.pdf', bbox_inches='tight', dpi=700)
plt.show()
```
