```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
df1 = pd.read_excel('../data/SLM_exp_02.xlsx',"Sheet3",skiprows=1)
```

```python
print(df1.columns)
```

```python
print(df1['X1'])
```

```python
delta_arr = df1['delta'][0:7:2].to_numpy()
# print(delta_arr)
psi_arr = df1['Psi'][0:7:2].to_numpy()
```

```python
grey_arr = np.array([0,64,128,255])
```

```python
plt.figure(figsize = (10,8))
plt.plot(grey_arr,delta_arr,'-o')
plt.xlabel("grey scale",math_fontfamily='cm',fontsize = 12)
plt.ylabel(r"$\delta$ (Linear retardance)",math_fontfamily='cm',fontsize = 12)
plt.title(r"Variation of $\delta$ over grey scale",fontsize = 16 , fontweight= 'bold')
plt.grid(True)
plt.savefig('Plot1.pdf', bbox_inches='tight', dpi=500)
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.plot(grey_arr,psi_arr,'-o')
plt.xlabel("grey scale",math_fontfamily='cm',fontsize = 12)
plt.ylabel(r"$\Psi$ (Linear retardance)",math_fontfamily='cm',fontsize = 12)
plt.title(r"Variation of $\Psi$ over grey scale",fontsize = 16 , fontweight= 'bold')
plt.grid(True)
plt.savefig('Plot2.pdf', bbox_inches='tight', dpi=500)
plt.show()
```
